# EDA (Exploratory Data Analysis)

# Gender

```python
df.columns
```

```python
ax=sb.countplot(x='Gender',data=df)
```

```python
ax=sb.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sb.barplot(x='Gender',y='Amount',data=sales_gen)
```

# Age

```python
df.columns
```

```python
sb.countplot(x='Age Group',data=df)
```

```python
sb.countplot(data=df,x='Age Group',hue='Gender')
```

```python
ax=sb.countplot(data=df,x='Age Group',hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

# Total Amount vs Age Group

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sb.barplot(x='Age Group',y='Amount',data=sales_age)
```

# State

```python
df.columns
```

```python
df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head()
```

```python
orders_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)

sb.barplot(data=orders_state,x='State',y='Orders')
```

```python
orders_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sb.set(rc={'figure.figsize':(15,5)})

sb.barplot(data=orders_state,x='State',y='Orders')
```

```python
amount_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sb.set(rc={'figure.figsize':(15,5)})

sb.barplot(data=amount_state,x='State',y='Amount')
```

# Marital Status

```python
ax=sb.countplot(x='Marital_Status',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sb.countplot(x='Marital_Status',data=df)

sb.set(rc={'figure.figsize':(5,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
marital_amount=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head()

sb.set(rc={'figure.figsize':(5,5)})

sb.barplot(data=marital_amount,x='Marital_Status',y='Amount',hue='Gender')
```

# Occupation

```python
sb.set(rc={'figure.figsize':(20,5)})

ax=sb.countplot(x='Occupation',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
occupation_purchasingpower=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sb.set(rc={'figure.figsize':(20,5)})

sb.barplot(data=occupation_purchasingpower,x='Occupation',y='Amount')
```

# Product Category

```python
sb.set(rc={'figure.figsize':(25,5)})

ax=sb.countplot(x='Product_Category',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
category_amount=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sb.set(rc={'figure.figsize':(25,5)})

sb.barplot(data=category_amount,x='Product_Category',y='Amount')
```

```python
ID_order=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sb.set(rc={'figure.figsize':(25,5)})

sb.barplot(data=ID_order,x='Product_ID',y='Orders')
```

# Another way of acheieving top 10 for above data is:
